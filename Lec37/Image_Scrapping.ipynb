```python
from urllib.request import urlopen
```

```python
url = 'https://unsplash.com/s/photos/aeroplane'
```

```python
data = urlopen(url)
```

```python
print(data)
```

```python
data_all = data.read()
```

```python
data_all
```

```python
from bs4 import BeautifulSoup
```

```python
data_all = BeautifulSoup(data_all, 'html')
```

```python
data_all
```

```python
all_img = data_all.find_all('img')
```

```python
all_img[0].attrs['src']
```

```python
# all_img
```

```python
import requests

for i in range(len(all_img)):
    try:
        img = requests.get(all_img[i].attrs['data-src'])
        print(all_img[i].attrs['data-src'])
    except:
        pass
    with open(f'C:/Users/bhatt/OneDrive/Desktop/whisper-main/DS_Live_Oct/Lec37/data-src/{i}.jpg', 'wb') as file:
        file.write(img.content)
```
