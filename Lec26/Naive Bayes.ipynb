```python
def prior_prob(Y_data, value):
    return np.sum(Y_data == value)/Y_data.shape[0]
```

```python
def likelihood_prob(X_train, Y_train, feat_col, feat_val, label):
    X_train = X_train[Y_train == label]
    numerator = np.sum(X_train[:, feat_col] == feat_val)
    denominator = np.sum(Y_train == label) # OR X_train.shape[0]
    return numerator/denominator
```

```python
def posterior_prob(Xtrain, Ytrain, Xtest):
    classes = np.unique(Ytrain)
    posterior = []
    
    for label in classes:
        likelihood = 1.0
        
        for i in range(Xtrain.shape[1]):
            cond_prob = likelihood_prob(Xtrain, Ytrain, i, Xtest[i], label)
            likelihood *= cond_prob
        
        prior = prior_prob(Ytrain, label)
        posterior.append(likelihood * prior)
        
    pred = np.argmax(posterior)
    
    return pred
```

```python
print(posterior_prob(train_X, train_Y, test_X[37]), test_Y[37])
```

```python
def accuracy(train_X, train_Y, test_X, test_Y):
    corr = 0
    
    for i in range(test_X.shape[0]):
        pred = posterior_prob(train_X, train_Y, test_X[i])
        if pred == test_Y[i]:
            corr += 1
    return corr/test_X.shape[0]
```

```python
accuracy(train_X, train_Y, test_X, test_Y)
```
