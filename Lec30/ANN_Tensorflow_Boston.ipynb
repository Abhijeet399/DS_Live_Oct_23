```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
```

```python
from keras.datasets import boston_housing
```

```python
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MinMaxScaler?
```

```python
min_max_scaler = MinMaxScaler()
```

```python
min_max_scaler.fit(train_X)
```

```python
train_X = min_max_scaler.fit_transform(train_X)
```

```python
min_max_scaler.fit(test_X)
test_X = min_max_scaler.fit_transform(test_X)
```

```python
min_max_scaler.fit_transform?
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential(name = 'Boston_Price_Prediction')
model.add(Dense(200, input_dim = train_X.shape[1], activation = 'relu', name = 'Layer_1'))
model.add(Dense(150, activation = 'relu', name = 'Layer_2'))
model.add(Dense(100, activation = 'relu', name = 'Layer_3'))
model.add(Dense(50, activation = 'relu', name = 'Layer_4'))
model.add(Dense(1, activation = 'linear', name = 'Layer_5'))
```

```python
model.compile(optimizer = 'Adam', loss = 'mse', metrics = ['mae'])
```

```python
model.summary()
```

```python
hist = model.fit(train_X, train_Y, batch_size = 40, epochs = 300, validation_split = 0.1)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
plt.plot(hist.history['mae'])
plt.plot(hist.history['val_mae'])
```

```python
preds = model.evaluate(test_X, test_Y)
```

```python
pred_result = model.predict(test_X)
```

```python
for i in range(len(pred_result)):
    print(pred_result[i], test_Y[i])
```
