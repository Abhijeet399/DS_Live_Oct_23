```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import classification_report, confusion_matrix
```

```python
raw_data = load_breast_cancer()
```

```python
raw_data
```

```python
print(raw_data.keys())
```

```python
df_data = pd.DataFrame(raw_data['data'], columns = raw_data['feature_names'])
df_target = pd.DataFrame(raw_data['target'], columns = ['Cancer'])
```

```python
df_data
```

```python
df_target
```

```python
df_data.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df_data, np.ravel(df_target), test_size = 0.2, shuffle = True)
```

```python
print(X_test.shape, Y_test.shape, X_train.shape, Y_train.shape)
```

```python
from sklearn import svm

svm_classifier = svm.SVC()
svm_classifier.fit(X_train, Y_train)
```

```python
preds = svm_classifier.predict(X_test)
```

```python
preds
```

```python
print(classification_report(Y_test, preds))
```

```python
from sklearn.model_selection import GridSearchCV

param_search = {
    'C': [0.1, 0.05, 1, 5, 0.001],
    'kernel': ['rbf', 'linear', 'sigmoid'], 
    'gamma': [1, 0.1, 0.01, 0.001]
}

grid = GridSearchCV(svm.SVC(), param_grid = param_search, refit = True, verbose = 0, cv = 5, return_train_score = True)
```

```python
grid.fit(X_train, Y_train)
```
