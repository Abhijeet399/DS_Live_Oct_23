```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D
```

```python
data, label = make_circles(n_samples = 1000, random_state = 20, noise = 0.05)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = label)
```

```python
print('Diameter along x:', max(data[:, 0]) - min(data[:, 0]))
print('Diameter along y:', max(data[:, 1]) - min(data[:, 1]))
```

```python
axis_1 = data[:, 0].reshape(-1, 1)
axis_2 = data[:, 1].reshape(-1, 1)
axis_3 = (axis_1**2) + (axis_2**2)

X = np.hstack((data, axis_3))
```

```python
X.shape
```

```python
axis = plt.subplot(projection = '3d')
axis.scatter(axis_1, axis_2, axis_3, c = label)
plt.show()
```

```python
from sklearn import svm

svm_classifier = svm.SVC(kernel = 'linear')
svm_classifier.fit(X, label)
```

```python
print(svm_classifier.score(X, label))
```

```python
print(svm_classifier.intercept_)
```

```python
svm_classifier.coef0
```
